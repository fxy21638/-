```python
# 导入必要的库
import numpy as np
import matplotlib.pyplot as plt
```

# 绘制 $\sin(x)$ 的图像
本 Notebook 演示如何使用 matplotlib 和 numpy 绘制 $\sin(x)$ 的图像。

## 生成 $x$ 数据
使用 numpy 生成从 $0$ 到 $2\pi$ 的等间隔数据点。

```python
# 生成 x 数据
x = np.linspace(0, 2 * np.pi, 100)  # 100 个点
```

## 计算 $y = \sin(x)$
对 $x$ 数据应用 numpy 的 sin 函数，得到对应的 $y$ 值。

```python
# 计算 y = sin(x)
y = np.sin(x)
```

## 绘制 $\sin(x)$ 图像
使用 matplotlib 绘制 $x$ 和 $y$ 的关系曲线，并添加标题和坐标轴标签。

```python
# 绘制 sin(x) 图像
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)')
plt.title('y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
